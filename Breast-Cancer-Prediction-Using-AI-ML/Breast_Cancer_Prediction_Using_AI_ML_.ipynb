IMPORTING THE DEPENDENCIES

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

DATA COLLECTION AND PROCESSING

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data , columns = breast_cancer_dataset.feature_names)
```

```python
# print the first five rows of the data frame
data_frame.head()
```

```python
# adding the target column to the data frame at the end
data_frame['diagnosis'] = breast_cancer_dataset.target
```

```python
#printing the last five rows of the data frame
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting information about the data
data_frame.info()
```

```python
# checking for the missing values in the dataset
data_frame.isnull().sum()
```

```python
# finding the statistical measures about the data
data_frame.describe()
```

Checking for distribution whether a cancer is malignant or benign
***1 --> Benign
0 --> Malignant ***

```python
data_frame['diagnosis'].value_counts()
```

```python
data_frame.groupby('diagnosis').mean()
```

DIVIDING THE TABLE INTO....

ALL THE INPUT FEATURES IN --> X

AND THE DIAGNOSIS(TARGET)  --> Y


```python
# for column the axis value is 1
# for row the axis value is 0

X = data_frame.drop(columns='diagnosis' , axis=1)
Y = data_frame['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

*SPLITING THE FEATURES INTO TRAINING DATA AND TESTING DATA*

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2 , random_state=2)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

Model training

**Logistic Regression Model**

```python
model = LogisticRegression()
```

```python
# training the logistic regression model using training data
model.fit(X_train , Y_train)
```

**MODEL EVALUATION**

*Accuracy Score*

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train , X_train_prediction)
```

```python
print('Accuracy on the training data = ' , training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test , X_test_prediction)
```

```python
print('Accuracy on the testing data = ' , test_data_accuracy)
```

**BUILDING A PREDICTIVE SYSTEM**

```python
input_data = (18.25,19.98,119.6,1040,0.09463,0.109,0.1127,0.074,0.1794,0.05742,0.4467,0.7732,3.18,53.91,0.004314,0.01382,0.02254,0.01039,0.01369,0.002179,22.88,27.66,153.2,1606,0.1442,0.2576,0.3784,0.1932,0.3063,0.08368)

#changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] == 0):
  print('The Breast cancer diagnosis :  Malignant')
else:
   print('The Breast cancer diagnosis :  Benign')
```
